```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
import time
import json

urls = [
    "https://www.nytimes.com/international/",
    "https://www.nytimes.com/international/section/us",
    "https://www.nytimes.com/international/section/world",
    "https://www.nytimes.com/international/section/politics",
    "https://www.nytimes.com/international/section/nyregion",
    "https://www.nytimes.com/international/section/education",
    "https://www.nytimes.com/international/section/sports",
    "https://www.nytimes.com/international/section/business",
    "https://www.nytimes.com/international/section/technology",
    "https://www.nytimes.com/international/section/science",
    "https://www.nytimes.com/section/weather",
    "https://www.nytimes.com/international/section/arts/dance",
    "https://www.nytimes.com/international/section/movies",
    "https://www.nytimes.com/international/section/arts/music",
    "https://www.nytimes.com/spotlight/pop-culture",
    "https://www.nytimes.com/international/section/arts/television",
    "https://www.nytimes.com/international/section/theater",
    "https://www.nytimes.com/international/section/arts/design",
    "https://www.nytimes.com/international/section/health",
    "https://www.nytimes.com/international/section/well",
    "https://www.nytimes.com/international/section/food",
    "https://www.nytimes.com/international/section/travel",
    "https://www.nytimes.com/international/section/style",
    "https://www.nytimes.com/international/section/fashion",
    "https://www.nytimes.com/international/section/realestate",
    "https://www.nytimes.com/international/section/opinion",
    "https://www.nytimes.com/international/section/opinion/columnists"
]

# Initialize WebDriver
driver = webdriver.Chrome()

all_links = []

try:
    for url in urls:
        driver.get(url)
        wait = WebDriverWait(driver, 10)

        # Scroll to the bottom of the page to load dynamic content
        last_height = driver.execute_script("return document.body.scrollHeight")
        while True:
            for scroll_position in range(last_height, last_height + 800, 80):  # Scroll incrementally
                driver.execute_script(f"window.scrollTo(0, {scroll_position});")
                time.sleep(0.05)  # Wait briefly for new content to load

            # Update the height after scrolling
            new_height = driver.execute_script("return document.body.scrollHeight")

            if new_height == last_height:  # Stop if no new content is loaded
                break
            last_height = new_height

        # Parse the page source with BeautifulSoup
        soup = BeautifulSoup(driver.page_source, 'html.parser')

        # Extract href links from <a> tags with the specified classes
        for class_name in ['css-9mylee', 'css-8hzhxf']:
            links = soup.find_all('a', class_=class_name)
            for link in links:
                href = link.get('href')
                if href and href.endswith('.html') and '/interactive/' not in href and '/games/' not in href and "/videos/" not in href:
                    # Convert relative URLs to absolute URLs
                    full_url = f"https://www.nytimes.com{href}" if href.startswith('/') else href
                    all_links.append(full_url)

finally:
    # Quit the WebDriver
    driver.quit()

# Save the collected links to a JSON file
with open('nytimes_links.json', 'w') as json_file:
    json.dump(all_links, json_file, indent=4)

print(f"Total links collected: {len(all_links)}")
```
